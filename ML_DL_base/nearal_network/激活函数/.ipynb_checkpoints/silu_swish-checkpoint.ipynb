silu(sigmoid-weighted linear unit)与dsilu(derivative of silu)的数学形式如下：
$$
f(x) = xsigmoid(x) \\
f(x) = sigmoid(x)(1 + x(1-sigmoid(x)))
$$

特点：<br/>
1)silu与dsilu的提出是用于增强学习中<br/>
2)silu的特点是在负值有一个全部最小值，可以充当正则器，防止梯度更新过大，因此具有自稳定（self-stabilizing）的特点（？）

swish的数学形式与silu一致，只是文章从不同的角度来阐释该激活函数

swish有三个主要的优点：<br/>
1)无上界。可以避免梯度饱和，从而避免梯度消失<br/>
2)有下界，非单调。有下界的特点可以为激活函数带来正则化的效果（一些比较大的负输入值被丢弃），非单调对于一些小的负输入值，可以产生负的输出（?）,同时降低模型对于参数初始化与学习率的敏感度（？）<br/>
3)平滑性。平滑性使激活函数的输出更加平滑(landscape)，从而使得loss函数的曲面更加平滑，有助于优化（？）
