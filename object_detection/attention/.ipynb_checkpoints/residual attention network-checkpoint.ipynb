residual attention network发表于17年的论文"Residual Attention Network for Image Classification"，是一种混合注意力模型。

residual attention network整体结构如下，由若干个串联的attention module组成，每个attention module负责处理不同的区域，如下第二图，底层的attention module负责抑制背景(sky)的特征，高层的attention module负责增强前景（热气球）的特征。每个attention module由两个分支组成，trunk branch与soft mask branch。其中，truck branch用于提取特征，与一般的卷积模块类似，soft mask branch用于产生注意力mask，产生的mask与trunck分支输出的特征图进行点乘操作，形成通道与空间上的混合注意力。soft mask branch如下图所示，是一种bottom-up top-down结构，类似于编码-解码结构。
