学习率衰减策略：

1.每隔一定的epoch衰减为原来的0.1，实现：scheduler_1 = torch.optim.lr_scheduler.MultiStepLR(optimizer, milestones=[30, 60， 90], gamma=0.1)

2.当loss不再降低时就衰减，实现：scheduler_1 = torch.optim.lr_scheduler。ReduceLROnPlateau(optimizer, mode='min', factor=0.1, patience=10)

3.cosine衰减（一般与warmup联合使用）：首次提出于论文"SGDR: STOCHASTIC GRADIENT DESCENT WITH
WARM RESTARTS"，计算公式如下：
$$
lr_{new} = eta_{min} + 0.5(lr_{init}-eta_{min})*(1+cos(\frac{epoch}{T_{max}}\pi))
$$
$其中,eta_{min}为最小学习率，T_{max}为cos周期的1/2，如T_{max}设为10，则前10个epoch学\\习率先从初始值衰减到最小值，然后后10个epoch从最小值增加到最大值。$<br/>
实现：CosineLR = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer_CosineLR, T_max=150, eta_min=0)
