swin transformer对transformer在图像中的应用提出了改进，指出swin transformer可以作为分类或密集型预测任务（检测/分割）的backbone

swin transformer主要改进：一是window-base multi-head attention，即多头注意力在窗口内计算，这使得计算量与输入图像尺寸的关系由平方转为线性。二是shifted window，即特征能够跨窗口交互，类似于卷积。三是patch merging, transformer输入与输出长度是一样的，因此无法得到多尺度特征，而patch merging类似于yolov5中的Focus结构，能够减小特征图尺寸。
