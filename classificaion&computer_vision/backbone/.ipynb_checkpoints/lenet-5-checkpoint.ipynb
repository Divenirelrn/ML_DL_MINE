lenet-5是98年由lecun等人提出的一种卷积网络，是最早的卷积网络在图像处理中的应用，它的提出是解决手写字体识别的问题。其结构图如下所示：

第一层是5x5的卷积层，然后是2x2的池化层，然后是5x5的卷积，再加一层2x2的池化，最后是一个5x5的卷积将特征图的维度调整至1x1，然后是一个120x84的全连接加一个84x10的全连接,其中，10是手写字体识别的类别数。

技术细节：<br/>
1)将激活层放在了池化层后面（现代的网络通常将激活层放在BN层之后）<br/>
2)卷积到全连接的过渡是采用的5x5的卷积，而不是直接变换维度<br/>
3)原文中的输出层并不是采用的softmax，而是RBF层（径向基核函数）

lenet-5可被看作卷积用于视觉图像处理的开山之作
