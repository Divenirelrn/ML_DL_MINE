elu(exponential linear units)的数学形式：
$$
f(x) =
 \begin{cases}
 x, \quad if \quad x>0\\
 \alpha(exp(x) - 1), \quad if \quad x<=0
 \end{cases}
$$

其导数形式为：
$$
f'(x) =
 \begin{cases}
 1, \quad if \quad x>0\\
 f(x) + \alpha, \quad if \quad x<=0
 \end{cases}
$$
$其中\alpha > 0$

1)有负值，可以将激活值的均值推向0(以0为中心)，使梯度更加接近于自然梯度（?），加速收敛<br/>
2)有下界(saturation region)，可以减少非激活值的方差（?），有助于网络学习到更加鲁棒与稳定的特征（?）
