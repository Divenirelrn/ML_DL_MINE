mixup发表于2018年的论文"mixup: BEYOND EMPIRICAL RISK MINIMIZATION"，提出一种数据集扩增方式，可有效提升分类与检测精度。此外，对于包含部分标注错误的数据集与对抗性样本有大幅提升。

mixup对一个mini-batch的随机两张图像进行混合，混合方式如下：
$$
\tilde{x} = \lambda x_i + (1-\lambda)x_j \\
\tilde{y} = \lambda y_i + (1-\lambda)y_j
$$
$其中，x_i,x_j为原始输入向量，y_i, y_j为one-hot标签,\lambda \sim Beta(\alpha, \alpha)\\分布，其中\alpha为Beta的参数向量，\alpha越大，两张图像混合比例越接近50%$

文章阐述了传统的方法是经验风险最小化，而mixup是相邻风险最小化（Vicinal Risk Minimization）

对于\alpha的值，作者在正常数据集设为0.2，0.4这样的值，在部分标注错误的数据集，设为8，32这样的值。
