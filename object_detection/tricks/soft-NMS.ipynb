soft-NMS于2017年在论文"Improving Object Detection With One Line of Code
"中提出，解决了NMS在处理高度重叠的两个检测框（均为FP）时，会去除score较低的检测框的问题，对于遮挡物体具有较好的效果，达到了优于NMS的效果。仓库见https://github.com/bharatsingh430/soft-nms

与NMS不同的是，soft-NMS并不会在每次迭代中都去除冗余框，而是会抑制score值，最后用阈值过滤统一去除冗余框。
