接受拒绝采样(accept-reject algorithm)是借助提议分布（Proposal Distribution）对目标分布进行采样的一种方法。

提议分布一般选择正态分布、均匀分布等常见分布

提议分布q(x)的概率密度函数需满足：<br/>
$$cq(x) >= p(x)$$
$其中c为人工设定的常数，c的取值应尽可能小，使得提议分布刚好能盖住目标分布$

对于从提议分布抽取的样本x,计算接受概率值：
$$
\alpha(x) = \frac{p(x)}{cq(x)}
$$
$然后生成一个服从均匀分布U(0,1)的随机数z，如果：$
$$
z <= \alpha(x)
$$
$则接受样本，否则拒绝样本$
