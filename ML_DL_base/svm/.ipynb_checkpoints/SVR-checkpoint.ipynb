SVR将SVM用于回归任务，决策边界变为回归曲线。传统的回归方式是直接计算预测值与标签之间的误差，预测值与标签完全相等时误差才为0.而SVR是可以容忍预测值与标签之间有一定的误差，只有差值大于误差时才计算损失，SVR的目标是使得间隔带以外的数据到回归曲线的距离最小

$如下图所示，以f(x)为中心，构造一个宽度为\epsilon的间隔带，落入间隔带中的样本就可判别为预测正确，\\落入间隔带之外的样本为支持向量$


$SVR的目标函数如下：$<br/>
$$
\min_{w,b}\frac{1}{2}||w||^2 + C\sum_{i=1}^{m}{l_{\epsilon}(f(x_i) - y_i)}
$$
$其中C为正则化系数，l_{\epsilon}为\epsilon-不敏感（\epsilon-insensitive loss）损失函数:$
$$
l_{\epsilon}(z) = 
 \begin{cases}
 0, if |z| <= \epsilon \\
 |z| - \epsilon, otherwise
 \end{cases}
$$

$引入松弛变量，则目标函数重写为：$
$$
 \begin{cases}
 \min_{w,b,\xi_i, \hat \xi_i}\frac{1}{2}||w||^2 + C\sum_{i=1}^{m}{(\xi_i + \hat \xi_i)}\\
 s.t. f(x_i) - y_i <= \epsilon + \xi+i \\
 y_i - f(x_i) <= \epsilon + \hat \epsilon_i \\
 \xi_i >= 0, \hat \xi_i >= 0, i= 1,2,...,m
 \end{cases}
$$
$由于SVM是希望样本点落入间隔带之外，而SVR是希望样本落入间隔带之内，\\所以需要有两个松弛变量，其中\xi_i是间隔带上面的数据到间隔带之间的垂直距离，\\ \hat \xi_i是间隔带下面的数据到间隔带之间的垂直距离，当数据落入间隔带上面时，\\ \xi_i>0,\hat \xi_i = 0;当数据落入间隔带下面时，\xi_i=0,\hat \xi_i > 0$

利用拉格朗日乘子法构造SVR的对偶问题，同时利用SMO与KKT条件可解得最优解。
