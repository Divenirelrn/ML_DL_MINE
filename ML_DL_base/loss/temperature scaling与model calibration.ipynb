temperature scaling(以下简称TS)的提出是在一篇名为"On Calibration of Modern Neural Network"的论文中

model calibration，即模型校准，是解决分类模型输出的各类置信度的均值与模型的准确率之间存在的不对齐（miscalibration）现象。如下图所示，左图是完美对齐的情况（LeNet on CIFAR-100），右图是非对齐的情况(ResNet110 on CIFAR-100)，右图中accuracy与置信度之间存在着明显的GAP。


在早期的分类模型中，非对齐现象较少出现，随着网络规模的发展，分类精度逐渐提升，但非对齐现象却越来越明显。

作者首先提出了一个衡量模型对齐问题的一个性能指标，推导如下：<br/>
$X为输入，Y为标签，K为类别数，\hat P 为置信度，则完美对齐（perfect calibration）定义为：$
$$
P(\hat Y = Y | \hat P = p) = p, \forall p \in [0,1]
$$
$即准确率等于置信度，例如，100个样本，置信度均为0.8，那么准确率也为0.8$

$对于上图的accuracy与confidence之间的曲线，我们把它分为M份，令B_{m}为\\置信度落入区间I_{m} = (\frac{m-1}{M}, \frac{m}{M})的样本的集合，则B_{m}的准确率可以定义为：$
$$
acc(B_{m}) = \frac{1}{|b_{m}|}\sum_{i \in B_m}{\pmb{1}(\hat y_i = y_i)}
$$
$例如，预测100个样本，对于每个样本输出一个类别及对应的置信度，落在0.0到0.2置信度区间\\的样本即为B_m，对于每一个样本，输出的类别不一定等于标签，因此\pmb{1}(\hat y_i = y_i)不一定等于0$<br/>
$B_m的平均置信度定义为：$<br/>
$$
conf(B_m) = \frac{1}{|B_m|}\sum_{i \in B_m}{\hat p_i}
$$
$其中\hat p_i为某个样本的置信度$<br/>
$那么我们定义ECE(Expected Calibration Error)为model calibration的衡量指标，定义为：$
$$
ECE = \sum_{m=1}^{M}{\frac{|B_m|}{n}|acc{B_m} - conf(B_m)|}
$$

最后，作者给出了解决miscalibration的方法，即temparature scaling。<br/>
TS是通过在验证集上最小化NLL，是在模型训练完成后，对模型进行后处理(post-processing)，不会影响模型的精度。具体做法是：<br/>
$$
logits_{new} = \frac{logits}{T}
$$
$即模型输出的logits（注意，是隐层输出的，而不是softmax输出的）乘以一个\frac{1}{T}，\\其中T > 1，一般初始化为1.5,是通过反向传播进行优化的参数，T称为temperature$
