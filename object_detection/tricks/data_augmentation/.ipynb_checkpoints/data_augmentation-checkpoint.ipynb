two-stage模型由于结构设计，许多data_aug手段效果不明显(ssd的观点)（论文"Bag of Freebies for Training Object Detection Neural Networks"的观点是two-stage模型自身有crop操作，因此不用randomcrop，而其他的几何形变如randomresize，randomexpand,randomhue,randomontrast等都使用），而one-stage模型能够受益于联合的数据集扩增手段，有时候能提高10几个点，甚至20多个点

数据集扩增旨在使得网络学习尺寸、长宽比、翻转、旋转、平移、color、光照等不变性。

submean：检测任务与分类任务不一样，分类一般对输入做一个normalization，而检测一般就减去均值进行中心化就可以了

faster RCNN只采用了horizontal flip数据扩增方式

yolov1采用了random scaling、random translation(二者均为20%),此外，还有random adjust exposure与saturation(二者factor均为1.5)<br/>
yolov1代码中的数据扩增方式包括：random_flip、randomscale、randomblur、randombrightness、randomhue、randomsaturation、randomshift、randomcrop，其中，randomscale指只对宽或高进行scale:


ssd采用的方式：1)随机裁剪：原图，裁剪一个与目标的iou值为0.1,0.3,0.5,0.7,0.9的区域，随机裁剪一个区域<br/>
这些裁剪后的patch固定到统一尺寸，然后horizontal flip,随后应用photo-metric distortion<br/>
2）随机expand<br/>
将原图缩小，放置在一个更大的背景上，生成更多小目标，有效提高小目标检测精度:


ssd代码中的数据扩增方式包括：submean、randomsaturation、randomhue、randomlightingnoise（随机交换通道）、randomcontrast、randombrightness、randomsamplecrop、expand、randommirror,此外photometricdistort包括randomcontrast、randomsaturation、randomhue：


yolov2采用的数据集扩增方式为：random crops,color shifting等。
