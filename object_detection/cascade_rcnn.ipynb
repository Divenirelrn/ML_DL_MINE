cascade rcnn的论文"Cascade R-CNN: Delving into High Quality Object Detection"发表于2017年，采用级联head来产生更精确的回归与分类结果。在rcnn-base的模型训练时，一般在第二阶段（fast rcnn），将iou0.5作为正负样本的阈值，这样输入到head的proposals质量较高。但对于测试阶段，这种单一的设置并不能给head输入高质量的proposals，这就造成了训练与测试阶段的不对齐。因此，作者采用多级级联的head，来提升proposals的质量，进而提升检测精度。
