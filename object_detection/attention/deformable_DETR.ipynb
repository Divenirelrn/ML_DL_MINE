detr存在两个主要缺点：<br/>
1)计算量太大<br/>
2)无法使用高分辨率特征图，导致小目标精度低（使用高分辨率特征图会导致计算量的增大）<br/>
3)收敛速度慢（detr初始化时，attention score几乎是均匀分布，因此，要学到有效的权重分布，需要很多次迭代，使得每层模型最终只关注特定的几个key）

deformable的主要贡献是用multi-scale deformable multi-head aelf-attention模块代替了encoder里的self-attention与decoder里的cross attention.其思想如上图，将deformable conv的思想用在self attention模块，在计算时，每一个query不必与所有的key计算输出attention score，而是与固定数量的输入作计算（sparse spatial sample），选择一个reference point，然后在此基础上用linear层计算偏移。
