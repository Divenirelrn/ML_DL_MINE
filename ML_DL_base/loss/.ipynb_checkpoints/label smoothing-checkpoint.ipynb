label smoothing起初是在一篇名为"Rething the inception architecture for computer vision"的论文中被提出，作为一种正则化分类层的手段。

传统的分类任务中，label的标注方式是one-hot，这种方式存在两个问题：<br/>
1)可能造成过拟合，one-hot方式迫使网络将所有的logits分配给对应于gt的那一维，不能保证泛化能力<br/>
2)会造成最大的logit与其他logits之间的差距特别大，造成模型对于自己的预测结果overconfident

基于此，作者提出了label smoothing，使得网络对于自己的预测结果less confident，这种情况下，标签形式变为：<br/>
$$
y_{k}^{LS} = (1 - \epsilon)y_k + \frac{\epsilon}{K}
$$
$其中，\epsilon为误差，一般取0.1，K为类别，实现时一般用K-1$

随后，在另一篇论文"When Does Label Smoothing Help"的论文中，对label smoothing的效果、适用场景进行了详细探讨

首先，作者可视化了加入label smoothing前后的类别效果：

可以看出，label smoothing使得分类后的类别之间更紧凑，减少了类内复杂度，同时增大了类间复杂度，从而提升网络的泛化能力，提升分类精度。

其次，作者阐述了label smoothing对于模型校准(model calibration)的贡献，在实验中取得了不亚于temperature scaling的效果。(有关model calibration的详细介绍，见temperature scaling)

最后，作者阐述了知识蒸馏中为何不适合用label smoothing，并且通过实验验证了加入label smoothing后，效果会下降。
