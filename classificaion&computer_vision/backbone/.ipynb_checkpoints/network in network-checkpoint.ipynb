network in network,简称NIN，于2013年提出，本文最大的贡献是1x1卷积的使用，首次用1x1卷积增加卷积操作的非线性（在1x1卷积后均有ReLU层），提升卷积操作的特征提取能力。

1x1卷积主要有两种作用：<br/>
1)提升网络的非线性<br/>
2)用于升降维（例如在bottleneck结构中，还有在inception结构中）

NIN有两个创新点：<br/>
1)在普通卷积之后，加入一个微型网络结构MLP，一般用两个1x1卷积实现，用于提升卷积操作的特征提取能力，成为MLPconv,其结构如下：

2)用全局池化层（global average pooling）代替全连接层，极大地减少了模型的参数量（VGG网络绝大多数参数位于全连接层）
