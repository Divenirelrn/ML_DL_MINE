cutmix于19年的论文"CutMix: Regularization Strategy to Train Strong Classifiers
with Localizable Features"中提出，是一种数据集扩增方法，文章指出，cutout会丢失全局信息，而mixup会引入不自然图像，因此cutmix克服了两者的缺点，带来了精度上的提升。

mixup公式如下：
$$
\tilde{x} = M \cdot x_A + (1-M) \cdot x_B \\
\tilde{y} = \lambda y_A + (1-\lambda)y_B
$$
$其中M为mask，\lambda与mixup一样，服从Beta(\alpha, \alpha)分布，文中\alpha设为1时效果最好$<br/>
mask M选取如下：
$$
r_x \sim Unif(0,W),r_w = W\sqrt{1-\lambda} \\
r_y \sim Unif(0,H),r_h = H\sqrt{1-\lambda}
$$
