shufflenet发表于2017年，核心思想是对组卷积进行channel shuffle操作，在提升精度的同时，构建了移动端的高效模型。

xception与resnext的小模型表现不佳，因为1x1卷积计算量较大，因此shufflenet利用组卷积来减少计算量。同时，为了克服组卷积在组之间上相互独立的缺点，提出了channel shuffle操作。

组卷积有一个缺点：输出特征图只来源于输入的某个分组，各个分组之间无法交流，如下图(a)所示。因此，可以对channel进行随机shuffle，如下图(b)所示，但这种操作随机性较大，因此可以对channel进行有规律的shuffle.
