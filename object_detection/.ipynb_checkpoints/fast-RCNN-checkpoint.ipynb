fast RCNN发表于2015年，其历史贡献是提出ROI pooling层，允许网络只对原始图像做一次特征提取，极大地减少了检测时间，同时提升了精度。此外，将检测结构改为fc层后接一个分类层与一个回归层，采用联合损失函数进行训练，精简了训练过程，同时避免了硬盘的使用（RCNN中需要先将roi特征保存至硬盘）。

ROI pooling可看做SPP的特殊版本，只有一个level(只有一个pooling尺寸)

fast RCNN的整体结构如下，检测流程为：首先原始图像经过CNN进行特征提取，同时采用ss算法生成2000候选框，然后将候选框映射到特征图，此后对于每一个候选框映射区域，经过ROI pooling生成一个固定尺寸的特征向量，输入后续fc层，最后将特征向量输入至分类层进行分类，输入至bbox 回归层进行位置修正。


训练过程如下：<br/>
1)mini-batch设为2，每张图像采64个roi(region of interst)，128个roi中，25%为正样本（与gt的iou大于0.5,剩余的为负样本,iou介于0.1至0.5，小于0.1的用于难例挖掘）<br/>
2)采用预训练模型进行初始化，将最后的pooling层换为roi pooling，同时将最后的fc层换为一个classifier与一个regressor<br/>
3）采用联合loss进行训练，联合loss形式如下：
$$
L(p,u,t^u,v) = L_{cls}(p,u) + \lambda [u>=1]L_{loc}(t^u, v)
$$
$其中，L_{cls}为分类损失，一般用cross entropy，L_{loc}用smooth L1 loss，为了增强loss\\对于噪声的鲁棒性，u>=1为损失函数，其中u代表类别，背景一般设为0类，因此u>=1表示\\只有非背景的roi才会计算定位损失，背景不计算。\lambda为平衡系数，作者在实验中采用\lambda=1$

bp算法计算roi pooling: bp算法用于roi pooling时，与max pooling相似，对于有重叠的roi,重叠部分像素的梯度值进行累加。

超参数设置：classifier的fc层设为g(0,0.01)，regressor的fc层初始化为g(0, 0.001)

尺寸不变性：一般有两种方法使得网络具有尺寸不变性，一个是数据集扩增，一个是多尺度训练。后面作者也用实验证明，深层网络擅长学习尺寸变化，multi-scale效果不明显，因此用单一尺寸训练与测试（尺寸设为600）。（注：此处提到图像金字塔，图像金字塔由于其计算量比较大，一般在测试时用）

利用SVD奇异值分解加速推理：在推理时全连接层成为性能瓶颈，因此用svd来加速推理：
$$
W \approx U\Sigma_t V^T
$$
$其中,U是uxt尺寸，\Sigma_t是txt，V是vxt。具体做法是将一个fc层分解为两个fc层串联的形式\\，注意不用激活函数，一个fc层的参数为\Sigma_t V^T，另一个参数为U，这样将fc的参数量从\\uv降至t(u+v)$

fast RCNN基于三种不同深度或宽度的网络，分别命名为model S，model M与model L.

参数冻结：首先作者通过实验证明，只微调fc层比全部微调精度低。随后作者指出不需要对所有的卷积层进行微调，虽然无精度上的区别，但可以加速网络收敛，作者在实验中冻结了1前几个conv层。

多任务学习提升性能：作者通过实验证明，multi-task训练能够带来精度上的提升。

更多数据是否有帮助：作者联合voc07与voc12进行训练，能够带来明显的精度提升，大数据集对于网络效果有较大的帮助

SVMs比softmax好吗：作者通过实验证明，用svm进行分类，效果还不如直接用softmax.

是否更多的proposal能带来更好的表现：作者通过实验证明，稀疏的roi提取比密集提取效果要好（作者此处提到AR，average recall，用于评估proposal的质量，待展开）
