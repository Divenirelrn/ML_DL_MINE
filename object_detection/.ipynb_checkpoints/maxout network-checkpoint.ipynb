Maxout Networks于2013年提出，其初衷是加强dropout的效果，采用maxout+dropout在当时得到state of art。后在检测领域用于multi-scale策略中，用于融合两个特征图。

maxout表达式如下，对于每一层来说，同时训练多组参数，然后对于每一个输出神经元，选择值最大的那个，示意图如下两图所示


maxout的优点有两个：<br/>
1.（为什么会提升dropout的表现）dropout相当于bagging策略（同时训练多个网络，最后取平均），但是一般在测试时，我们只对使用dropout层的后续权重乘以drop_ratio，这是一种近似平均，而不是真正的平均，这种方式对于线性激活函数来说比较精确，但对于非线性激活函数来说就不精确。而maxout是一个分段线性函数，因此得到的结果更精确。<br/>
2.对优化更有利。相比于ReLU，maxout的表达式中没有0值，因此不会带来死亡神经元问题，优化效果更好。

缺点：maxout最大的缺点是参数量呈k倍增长，其中k为训练k组参数。
