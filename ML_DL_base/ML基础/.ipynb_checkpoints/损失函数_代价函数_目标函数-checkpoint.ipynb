损失函数（代价函数, cost function, loss）: L(Y, f(x)) = (Y - f(x))^2

风险函数（经验风险）：损失函数的期望(关于训练集的平均损失，即经验风险)，即：<br/>
$$
\frac{1}{N}\sum_{i=1}^{N}{L(y_i, f(x_i))}
$$
最小化经验风险，即经验风险最小化，下图中，图1的经验风险最大，图3的经验风险最小

结构风险：J(f)专门用来度量模型的复杂度，即正则化项，最小化结构风险即结构风险最小化<br/>
上图中，图1的结构风险最小，图3的结构风险最大

目标函数：
$$
min\frac{1}{N}\sum_{i=1}^{N}{L(y_i, f(x_i)) + \lambda J(f)}
$$
即同时最小化经验风险与结构风险
