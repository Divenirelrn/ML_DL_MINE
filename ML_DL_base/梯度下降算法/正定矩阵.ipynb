矩阵A对应的二次型$X^TAX$，对于任意非0X,其值均大于0，则为正定二次型，矩阵A为正定矩阵

A正定的判定：<br/>
(1)矩阵A的n个特征值均大于0<br/>
(2)存在可逆矩阵P,使得$A = P^TP$<br/>
(3)若A为正定矩阵，则$A^T$也为正定矩阵<br/>
(4)矩阵A的顺序主子式均为正

特征值与特征向量：<br/>
$A\alpha = \lambda \alpha$<br/>
$|\lambda E - A| = 0$

主子式：<br/>
在n 阶行列式中，选取行号（如 1、3、7行），再选取与行号相同的列号（1、3、7 列），则行数和列数都为i个的行列式即为n阶行列式的i阶主子式，也可以说由上述选取的行列交汇处的元素所组成的新的行列式 就称为“n 阶行列式的一个 i 阶主子式”。<br/>
