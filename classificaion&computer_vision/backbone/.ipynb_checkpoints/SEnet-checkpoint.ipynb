Senet的历史贡献是关注通道间的关系，提出一个新的模块squeeze-and-extation(SE) 模块。其中心思想是计算特征图channel的重要程度，计算出权重向量，然后每个维度乘以相应的权重。这一过程称为特征校准（feature recalibration），使模型抑制信息量少的channel，强调包含信息多的channel。具体做法是先squeeze，将特征图尺寸缩放至1x1，然后通过两个全连接层计算权重向量。

SE模块结构见下图。首先是squeeze操作，利用特征图全局感受野来产生一个embedding。然后是excitation操作，采用自门控（self-gating）机制来产生权重集合。


SE模块可以直接嵌入其它网络结构，但在网络中的前面层，它提取的是较通用的特征，但在后面层，它提取的是针对类的较专有的特征。

Squeeze:global information embedding。利用global average pooling实现。<br/>
Excitation:adaptive recalibration.excitation操作旨在学习通道间的非互斥关系，通过非线性变换与全连接层实现，其中全连接层主要用于升降维。具体操作如下图所示。
