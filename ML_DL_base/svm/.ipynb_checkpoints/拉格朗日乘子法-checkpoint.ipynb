对于包含等式约束条件的优化问题：<br/>
$$ 
minmax f(x) \\
s.t. g(x) = 0
$$

$可以构造以下方程组来解决：$
$$
 \begin{cases}
 \nabla f(x) = \nabla g(x) \\
 g = 0
 \end{cases}
$$


其中，第一个式子的依据是：<br/>
目标函数的梯度与等值线的切线方向垂直<br/>
在极小点处，目标函数的梯度与约束的梯度平行：


拉格朗日方程为：$L(x, \lambda) = f(x) + \lambda g(x),其中\lambda >= 0$<br/>
$求解\frac{\partial{L}}{\partial{x}}=0与\frac{\partial{L}}{\partial{\lambda}}=0等价于求解上述方程组$
