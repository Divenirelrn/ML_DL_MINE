Alexnet发表于2012年的论文"ImageNet Classification with Deep Convolutional
Neural Networks"，作者是Alex与Hinton等人，将卷积网络用于图像识别，在2012年的Imagenet(ILSVRC-2012)图像识别任务中，取得第一名的成绩，其历史贡献是使得人们真正开始关注深度学习及其在图像、NLP等领域的应用。

作者首先介绍了Imagenet，说明了高像素大数据集对于机器学习方法的重要性。又介绍了卷积相对于全连接层的优势，及GPU的发展支持大规模神经网络的训练。

2)local respomse Normalization。该方法首次在Alexnet中被用到，后面经过多人测试发现效果并不太好，因此现在被弃用。作为一种Normalization方法，其原理如下：<br/>
$令a_{x,y}^i为特征图中第i个channel在位置(x,y)处的激活值（after ReLU），经过LRN之后的值为：$
$$
b_{x,y}^i = a_{x,y}^i / (k + \alpha \sum_{j=max(0, i-n/2)}^{min(N-1, i+n/2)}{(a_{x,y}^j)^2})^{\beta}
$$
3)overlapping pooling。overlapping pooling是重叠区域的池化层，若池化层filter size为z,slide步长为s,若s<z，则为重叠下采样。经作者试验，重叠下采样具有一点防止过拟合的作用
