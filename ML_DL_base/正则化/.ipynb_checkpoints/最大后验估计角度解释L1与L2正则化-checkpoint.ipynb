最大似然估计（MLE）：<br/>
$p(x_1, x_2, ...|\theta) = p(x_1|\theta)p(x_2|\theta)... = \prod_{i=1}^{m}{p(x_i|\theta)}$<br/>
最大化似然函数即为最小化loss函数

最大后验概率估计(MAE)：<br/>
$p(\theta|x) = \frac{p(x|\theta)p(\theta)}{p(x)} = \frac{p(x|\theta)p(\theta)}{\sum_{i=1}^{n}{p(x_i|\theta)p(x_i)}}$

$p(\theta|x)的分子的两项，第一项是MLE,第二项是先验概率，是事先假设参数服从某种分布$

$max_\theta p(\theta|x) = max_\theta p(x|\theta)p(\theta) = min_\theta [-p(x|\theta)p(\theta)]$

$若假设参数服从拉普拉斯分布，即p(\theta) = \frac{1}{2}e^{-|\theta|}，如下方红线所示，则min_\theta [-ln(p(x|\theta)p(\theta))] = min_\theta [-ln(p(x|\theta)) - \frac{1}{2}|\theta|] ，即加了L1正则化项的loss函数。\\若假设参数服从高斯分布，即p(\theta) = \frac{1}{\sqrt{2\pi}}e^{-\frac{\theta^2}{2}}，如下方蓝线所示，则min_\theta [-ln(p(x|\theta)p(\theta))] = min_\theta [-ln(p(x|\theta)) - \frac{1}{2\sqrt{2\pi}}\theta^2] ，即加了L2正则化项的loss函数。$


$若假设参数服从拉普拉斯分布，从其图像可以看出，参数取值为0的概率比较高，因此会有一些参数值为0 \\ 若假设参数服从高斯分布，从其图像可以看出，参数靠近0的概率较高，因此参数值普遍较小$

$注意：拉普拉斯原形式：p(\theta) = \frac{1}{2\lambda}e^{-\frac{|\theta-\mu|}{\lambda}}$<br/>
$高斯分布原形式：p(\theta) = \frac{1}{\sqrt{2\pi}}e^{-\frac{(\theta-\mu)^2}{2\delta^2}}$
