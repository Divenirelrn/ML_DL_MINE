知识蒸馏是模型压缩的三大方法之一，模型压缩主要包括剪枝、蒸馏、量化。

大模型输出一个分类结果，但同时也输出其它错的分类结果（概率），这些概率可能很低，但其中一些的概率比其它的大得多。例如，一张car的分类图像，输出car的概率值最大，但输出truck的概率要比输出carrot的概率大得多。

蒸馏得来的模型，泛化性能与teacher 模型相似，比用常规方法训练一个小模型好的多。
将大模型的泛化性能迁移到小模型上，一个好的做法是将大模型生成的类别概率作为"soft targets"，对于ensemble的模型来说，可以将他们的均值作为软标签。软标签的熵越大，能提供的信息越多。这样，小模型就能用更少的数据与更大的learning rate.

但在一些情况下，一些模型输出其它类的概率很小，如mnist数据集，一张2的图像，可能输出3的概率是10^(-6)，输出7的概率是10^(-9)，这样效果更接近于硬标签，不利于标签的平滑。为此，作者提出蒸馏的方式，to raise the temperature of the final softmax until the cumbersome model produces a suitably soft set of targets，即使得大模型生成的标签更平滑。

知识蒸馏的步骤是：<br/>
1)训练teacher网络<br/>
2)训练student网络，用teacher网络生成soft target，计算soft loss,与原标签的hard target一起组成loss，并进行梯度反传。

softmax_t是关键方法，teacher网络用它来产生soft target，student网络用同样的T来产生输出
