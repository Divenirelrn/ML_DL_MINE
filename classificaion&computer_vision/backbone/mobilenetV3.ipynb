mobilenetV3发表于2019年，文章标题为"searching for mobilenetV3"，是将网络搜索技术（NAS）与模型设计相结合而产生的模型。其中，模型设计是在mobilenetV2的基础上进行。mobilenetV3取得了精度与速度均优于mobilenet V2的成绩。

mobilenetV3包含mobilenetV3-small与mobilenetV3-large两个结构。

mobilenetV1采用深度可分离卷积。mobilenetV2采用inverted residual block with linear bottleneck, MnasNet在mobilenetV2的基础上引入轻量级的注意力机制（由SEnet得来）。mobilenetV3用以上所有层的组合，以构建一个最有效的模型。mobilenet v3的基本结构及与mobilenetV2的对比如下图所示。

网络设计<br/>
对计算量大的层重新设计：mobilenetV2的最后一层是1x1卷积，将该层及之前的bottleneck去除，使得avgpooling的输入变为1x1，而不是原来的7x7，如图5所示。这样延时减少7ms，相当于提升11%。


非线性：作者用swish代替ReLU，换来性能上的提升，但却带来计算上的增加，主要原因是sigmoid涉及指数运算，在移动设备上计算量较大。为了解决这个问题，作者采用h-sigmoid（hard-sigmoid）来代替sigmoid，从而用h-swish代替swish，在保持精度的基础上减少计算量。由于h-sigmoid不涉及指数运算，因此计算量较小。H-sigmoid数学表达式为$\frac{ReLU6(x+3)}{6}$，因此h-swish的数学表达式如下：<br/>
$$
h-swish[x] = x\frac{ReLU6(x+3)}{6}
$$
