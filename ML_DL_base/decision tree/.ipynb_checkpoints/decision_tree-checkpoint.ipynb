决策树是一种分类算法，其思想向信息增益下降最快的方向构建决策树。决策树的类型包括ID3算法与C4.5算法，其中C4.5算法在实践中最为常用。

决策树应注意平衡性，各分支结点数量要大致相等。

决策树的构建原则是减小不确定性，信息熵用于不确定性的衡量：
