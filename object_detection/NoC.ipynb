"Object Detection Networks on Convolutional Feature Maps"论文发表于2017年，其历史贡献是探索了在目标检测架构中，采用深层的、包含conv层的classification sub network（NoC），实验表明，用更深层的（前提是不要过拟合）、包含conv的NoC能够提升检测效果（提升定位的准确性）。

进一步，作者用多尺度特征图结合maxout构建分类子网络，进一步提升性能。这里的multi-scale是对于每一个候选区域，随机选取两个相邻尺寸的特征图进行投影。


fine-tuning NoC：作者发现，对分类子网络（NoC)imagenet预训练模型进行初始化能够大幅提升精度.

roi pooling提前：作者发现，采用resnet的faster RCNN，精度不如采用vgg的faster RCNN，作者猜想是VGG下采样倍数为16，而resnet下采样倍数为32,因此特征图尺寸较小，不利于检测。因此，在将NoC应用到resnet-faster RCNN中时，将roi pooling layer提前两层，这样在roi pooling之前，下采样倍数为16，被提前的两层conv用作NoC的conv层使用
