对比学习的发展历程主要分为百花齐放、CV双雄、不用负样本、transformer四个阶段

百花齐放：

百花齐放主要发生在18年-19年中，主要工作有InstDisc（即提出instance discrimitive）、CPC、CMC等。

该阶段的特点是代理任务多种多样，有预测类型的，有生成类型的，有数据增强的，有multi-view的。

CV双雄：

该阶段的代表性工作是moco v1,moco v2,simclr v1,simclr v2,此外还有swav.

其中swav用聚类中心来做对比学习，其工作中，multi-crop方法被广为应用，成为对比学习的标配。

不用负样本：

经过CV双雄阶段，对比学习的模式基本被固定下来，loss用infoNCE，网络结构上用encoder+projection head, 都用比较强的数据增强，都用动量编码器，都训练更长时间的epoch。但都需要负样本，因此诞生了一些不用负样本，而是基于预测的方法。

负样本是一种约束，如果没有负样本，容易使得网络学得捷径解，对于所有的输入有相似的输出，造成网络坍塌。

代表性的工作有BYOL（Bootstrap your own latent,latent为特征的意思）、SimSiam，它们都是多了一层预测层，使得预测特征向量与正样本的特征向量尽量一致。
