GHM提出于2018年的论文"Gradient Harmonized Single-stage Detector"，提出一种新的loss函数，解决hard exsample的问题，并应用于检测的分类头，为了验证GHM的效果，还设计了回归loss。最终达到了由于focal loss的效果。

focal loss的设计目的在于抑制易分样本对于loss的贡献，如下图中间所示。而GHM通过对已收敛模型的梯度norm的分析（如下图左所示），认为决定梯度方向的不仅是大量易分样本（下图左左），还有部分样本也影响梯度更新方向（下图左右），这部分样本成为outliers，是非常难分的样本，因此也应该抑制这部分样本的loss(下图中红线)


binary cross entropy loss可表示为：
$$
L_{CE}(p, p^*) = 
\begin {cases}
-log(p), if p^*=1 \\
-log(1-p), if p^* = 0
\end {cases}
$$
则其导数可表示为：
$$
\frac{\partial{L_{CE}}}{\partial{x}} = 
\begin {cases}
p-1, if p^*=1 \\
p,if p^*=0 \\
\end {cases}
= p - p^*
$$
我们定义梯度norm如下：
$$
g = |p-p^*| = 
\begin {cases}
1-p, if p^*=1 \\
p, if p^* = 0
\end {cases}
$$
g表示了一个样本的属性（难样本或易分样本）

则进一步，我们表示梯度密度如下：
$$
GD(g) = \frac{1}{l_{\epsilon}(g)}\sum_{k=1}^{N}{\delta_{\epsilon}(g_k, g)}
$$
$其中，g_k是第k个样本的梯度范数，且：$
$$
\delta_{\epsilon}(x,y) = 
\begin {cases}
1, if y-\frac{\epsilon}{2} <= x < y + \frac{\epsilon}{2} \\
o, otherwise
\end {cases}
$$
$$
l_{\epsilon}(g) = min(g+\frac{\epsilon}{2}, 1) - max(g-\frac{\epsilon}{2}, 0)
$$
$梯度密度指示了处在以g为中心、\epsilon为长度的区域里的样本数量，且被该区域的有效长度规范化$

基于此，我们可定义梯度密度平衡参数（gradient density harmonizing parameter）：
$$
\beta_i = \frac{N}{GD(g_i)}
$$
$其中，N为样本数量，\beta_i可重写为：$
$$
\beta_i = \frac{1}{GD(g_i)/N}
$$
$每个样本乘以自己的\beta_i，从而梯度密度较大的样本受到抑制$

GHM-C loss<br/>
将GHM loss应用于分类头：
$$
L_{GHM-C} = \frac{1}{N}\sum_{i=1}^{N}{\beta_iL{CE}(p_i, p_i^*)} = \sum_{i=1}^{N}{\frac{L_{CE}(p_i, p_i^*)}{GD(g_i)}}
$$

Unit Region Approximation<br/>
计算所有样本的梯度密度，其时间复杂度为$o(N^2)$，为了简化这种计算，作者提出用一种近似的方法来求。<br/>
$将g平均划分为若干单元，每个单元长度为\epsilon，则一共有M=\frac{1}{\epsilon}个单元，\\令r_j为第j个单元，则r_j = [(j-1)\epsilon, j\epsilon)，令R_j为r_j中的样本数量。定义\\ind(g)=t, s.t.(t-1)\epsilon <= g < t\epsilon，为下标函数，指示g所处的单元。基于此，\\可定义近似梯度密度函数：$
$$
\hat{GD}(g) = \frac{R_{ind(g)}}{\epsilon}=R_{ind(g)}M
$$
$进而可得近似梯度密度平衡参数：$
$$
\hat \beta_i = \frac{N}{\hat{GD}(g_i)}
$$
$最终GHM-C可重写为：$
$$
\hat L_{GHM-C} = \frac{1}{N}\sum_{i=1}^{N}{\hat \beta_i L_{CE}(p_i, p_i^*)} = \sum_{i=1}^{N}{\frac{L{CE}(p_i, p_i^*)}{\hat{GD}(g_i)}}
$$
$该计算方式时间复杂度为o(MN)$

EMA<br/>
基于mini-batch的方法会带来统计上的不准确，带来噪声。EMA(exponential moving average)被广泛使用来解决这个问题，如SGD with momentum，BN。作者将EMA应用于梯度密度的计算，以使其更平滑，对异常数据不敏感：
$$
S_j^{(t)} = \alpha S_{j}^{(t-1)} + (1-\alpha)R_j^{(t)}
$$
则：
$$
\hat{GD}(g) = \frac{S_{ind(g)}}{\epsilon} = S_{ind(g)}M
$$

GHM-R loss<br/>
$t=(t_x, t_y, t_w, t_h)为预测偏移，t^* = (t_x^*, t_y^*, t_w^*, t_h^*)为标签，smooth L1 loss如下：$
$$
L_{reg} = \sum_{i \in \{x,y,w,h\}}{SL_1(t_i, t_i^*)}
$$
其中，
$$
SL_1(d) = 
\begin {cases}
\frac{d^2}{2\delta}, if |d| <= \delta \\
|d| - \frac{\delta}{2}, otherwise
\end {cases}
$$
其导数为：
$$
\frac{\partial{SL_1}}{\partial{t_i}} = \frac{\partial{SL_1}}{\partial{d}} = 
\begin {cases}
\frac{d}{\delta}, if |d| <= \delta \\
sgn(d), otherwise
\end {cases}
$$
$其中sgn为sign函数$

为了将GHM应用于regression函数，首先将smooth l1 Loss修改为：
$$
ASL_1(d) = \sqrt{d^2 + \mu^2} - \mu
$$
其导数为：
$$
\frac{\partial{ASL_1}}{\partial{d}} = \frac{d}{\sqrt{d^2 + \mu^2}}
$$
$文中\mu设为0.02$<br/>
