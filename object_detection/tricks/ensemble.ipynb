通用的ensemble method for object detection:该论文提出一种通用的目标检测ensemble策略，独立于模型结构的设计，并且将该方法应用于TTA(test time augmentation)，此外，还使用了数据蒸馏与模型蒸馏技术。下面分别介绍：

1.model ensemble<br/>
不同的模型产生不同的检测框，每个检测框包含位置、类别与score值，该结构是[model1 results,model2 results,...,modelm results]。首先，把初始的检测框展平，所有的检测框放在一个列表里；然后对这些检测框进行分组，把相同类别且iou值大于0.5的放在同一组里，这样每一组的检测框只关注于某个区域；然后关于每个组里是否包含物体(该组是否被保留)，有以下3种投票策略：<br/>
(1)Affirmative.就是只要一个组的length大于0，也就是说只要有一个检测器说该区域有物体，那么这个组就被保留；<br/>
(2)consensus.一个组的length大于m/2才得以保留，其中m为检测器个数，也就是说，只有大部分检测器同意这个区域有物体，那么这个group才得以保留；<br/>
(3)unanimous.一个组的length等于m，这个组才得以保留，也就是说，所有检测器同意这个区域有物体，那么这个组才得以保留。<br/>
最后，对每个保留下来的组分别实行NMS。<br/>
三种投票策略的效果对于不同的原始模型（base model）会产生不同的提升效果，当原始模型包含较少的FP时，即原始模型大概检测正确时，Affiemative效果较好，反之unanimous效果较好，而consensus是两者的tradeoff.

注：个人觉得这个地方有些不合理，应该对每个检测框做一个标记，标志它来自哪个检测器，这样投票策略才合理

2.TTA<br/>
DATA augmentation在分类任务中表现较好，在检测任务中表现较差，因为像traslation、crop等方法会改变目标物体的位置信息，因此一般只用color jittering等不改变位置信息的方法，而本文对data augmentation的位置进行了修正。对原始的图像做data augmentation，然后分别得到各自的检测结果，对于horizontal flip，在做完检测后，还要将图像与检测框恢复到原始图像的位置（object tranformation），即再flip回来，然后对三种不同的检测结果应用ensemble。

WBF:WBF的提出源自21年的论文"Weighted boxes fusion: Ensembling boxes from different object detection models"，是迄今为止效果最好的object detection领域ensemble方法。WBF应用于ensemble与TTA。均达到较好的效果，在ensemble的应用中，最高可提升10个点。其仓库地址见于：https://github.com/ZFTurbo/Weighted-Boxes-Fusion

WBF(weighted boxes fusion)具体流程如下：<br/>
我们有N个模型得到的bboxes，或同一个模型用TTA得到的N个预测结果
1)所有的bbox加入list B, 并按cls score排序<br/>
2)定义L与F列表，其中L用于存放类别相同、iou>THR(iou thres)的组，每个组是一个list，F与L长度相等，用于存放对应组的fusion结果<br/>
3)for box in B:<br/>
在F中找与其iou大于THR(i.e. 0.55)的box_in_F，<br/>
若找到了，将该box加入box_in_F对应的L的组中<br/>
若未找到，将该box append进F与L列表中<br/>
每次F在pos位置添加进新的box，利用L[pos]的T个boxes重新计算F[pos]的score值及其坐标：
$$
C = \frac{\sum_{i=1}^{T}{C_i}}{T} \\
X1,2 = \frac{\sum_{i=1}^{T}{C_i*X1,2_i}}{\sum_{i=1}^{T}{C_i}} \\
Y1,2 = \frac{\sum_{i=1}^{T}{C_i*Y1,2_i}}{\sum_{i=1}^{T}{C_i}}
$$
$注：权值也可用非线性的，如C^2, \sqrt{C}等$<br/>
这样，score值高的box对于坐标值的贡献要高于score值低的<br/>
4)当遍历完成后，re-scale B中box的score值（如果L一个组的box数量较少，也就是说只有少部分model同意该区域有box，那么F对应位置的score就要降低）：<br/>
$$
C = C*\frac{min(T,N)}{N} \\
or \quad C = C * \frac{T}{N}
$$
两种方法差别不大
