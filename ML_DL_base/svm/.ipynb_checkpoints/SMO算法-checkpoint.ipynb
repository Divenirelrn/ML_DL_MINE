SMO(sequentual minimal optimize)用于求解SVM的对偶问题，求出$\lambda$,即可得到原问题的最优解。<br/>
SVM(软间隔)的目标函数为：
$$
\min_{w,b}||w||^2 + C\sum_{i=1}^{m}{\xi_i} \\
s.t. y_i(w^T\Phi(x_i) + b) >= 1 - \xi_i \\
\xi_i >= 0
$$
拉格朗日函数为：<br/>
$L(w, b, \lambda) = \frac{1}{2}||w||^2 + C\sum_{i=1}^{m}{\xi_i} + \sum_{i=1}^{m}{\lambda_i (1- \xi_i - y_i(w^T\Phi(x_i) + b))} - \sum_{i=1}^{m}{\mu_i \xi_i}$<br/>
SVM（软间隔）的对偶问题为：
$$
 \begin{cases}
 \min_{\lambda} [\frac{1}{2}\sum_{i=1}^{m}{\sum_{j=1}^{m}{\lambda_i\lambda_jy_iy_jK_{ij}}} - \sum_{i=1}^{m}{\lambda_i}]\\
 s.t. \sum_{i=1}^{m}{\lambda_iy_i} = 0 \\
 0 <= \lambda_i <= C
 \end{cases}
$$

SMO的整体思路与坐标下降法类似，每次选择最不满足kkt条件的一对$\lambda$进行优化，直到所有的$\lambda$满足kkt条件<br/>
$为什么每次选择一对，而不是一个：若选择一个，则通过约束\sum_{i=1}^{m}{\lambda_iy_i} = 0，那么该\lambda可由其他的\lambda表示，无法达到优化的效果$

SVM的KKT条件要求（其中的x_i可以为$\Phi(x_i)$）：
$$
\begin{cases}
 y_i(w^Tx_i + b) - 1 + \xi_i >= 0 \\
 \xi_i >= 0 \\
 \lambda_i (1- \xi_i - y_i(w^Tx_i + b)) = 0 \\
 \mu_i \xi_i = 0 \\
 \lambda_i >= 0, \mu_i >= 0
\end{cases}
$$

$\frac{\partial{L}}{\partial{\xi_i}} = 0 \rightarrow \mu_i + \lambda_i = C$

$由KKT条件与\mu_i + \lambda_i = C可以推得：$<br/>
$\lambda_i = 0 时，\mu_i = C, 则\xi_i = 0, 则 y_i(w^Tx_i + b) >= 1$<br/>
$0 < \lambda_i < C 时, \mu_i = C - \lambda_i > 0, 因此\xi_i = 0, 又 \lambda_i (1- \xi_i - y_i(w^Tx_i + b)) = 0， 则y_i(w^Tx_i + b) = 1$<br/>
$\lambda_i = C时，\mu_i = 0, \xi_i >= 0, 又 \lambda_i (1- \xi_i - y_i(w^Tx_i + b)) = 0,则1- \xi_i - y_i(w^Tx_i + b) = 0， y_i(w^Tx_i + b) = 1- \xi_i，因此y_i(w^Tx_i + b) <= 1$

$挑选一对不满足kkt条件的\lambda时，首先挑选一个不满足kkt条件的\lambda,所谓不满足kkt条件的\lambda，就是某个样本的点对应的\lambda,在上面三种情况中与kkt条件相差最远的，\\
例如\lambda_i = 0, 则此时y_i(w^Tx_i + b)应该大于等于1，若不满足，计算二者的差距$

$第二个\lambda的选取原则：选取w^Tx_i + b与y_i差距最大的样本点对应的\lambda,计算公式为：$
$$
|E_1 - E_2|
$$
$其中，E_1 = (w^Tx_1 + b) - y_1 = \sum_{i=3}^{m}{\lambda_iy_i}K_{i1} + \lambda_1^{old}y_1+K_{11} + \lambda_2^{old}y_2K_{21}+b^{old} - y_1, 其中K_{i1} = x_ix_1,K_{11} = x_1x_1,K_{21} = x_2x_1$<br/>
$E_2 = (w^Tx_2 + b) - y_2 = \sum_{i=3}^{m}{\lambda_iy_i}K_{i2} + \lambda_1^{old}y_1+K_{12} + \lambda_2^{old}y_2K_{22}+b^{old} - y_2, 其中K_{i2} = x_ix_2,K_{12} = x_1x_2,K_{22} = x_2x_2$<br/>
$E_1与E_2为预测值与标签之间的差距，当E_1>0时，选取最小的E_i作为E_2;当E_1<0时，选取最大的E_i作为E_2$<br/>
$不管是最小的还是最大的E_i作为E_2，都是选择预测值与真实值差距最大的E_i$

$选取好了两个要优化的\lambda_i与\lambda_j之后，将对偶问题转化为只包含\lambda_i与\lambda_j的问题：$
$$
\begin{cases}
min_{\lambda_1, \lambda_2}[ \frac{1}{2}K_{11}\lambda_1^2 + \frac{1}{2}K_{22}\lambda_2^2 + y_1y_2K_{12}\lambda_1\lambda_2 - (\lambda_1 + \lambda_2) + y_1\lambda_1\sum_{i=3}^{m}{\lambda_iy_iK_{i1}} + y_2\lambda_2\sum_{i=3}^{m}{\lambda_iy_iK_{i2}}] \\
s.t. \lambda_1y_1 + \lambda_2y_2 = -\sum_{i=3}^{m}{\lambda_iy_i} = \zeta \\
0 <= \lambda_i <= C, i=1,2
\end{cases}
$$

$将\lambda_1用\lambda_2进行表示：\lambda_1 = \frac{\zeta - \lambda_2y_2}{y_1}$<br/>
$则优化问题变为一个只包含\lambda_2的二次函数（凸函数，开口向上），对\lambda_2进行求导，并令导数等于0即可得最小值，最小值一定在\lambda_1y_1 + \lambda_2y_2 = \zeta这条直线上$

$又0 <= \lambda_i <= C, 所以若最小值落在矩形区域内，则最小值即为最优解，若落在矩形区域外，则需要对最小值进行裁剪：$<br/>
$ \lambda_2^{new}的取值范围为L <= \lambda_2^{new} <= H，当\lambda_2^{new} < L时，将其值设为L, 当\lambda_2^{new} > H时，将其值设为H,其余情况正常取值，L与H为取值边界，计算如下： $<br/>
$对于y_1 \ne y_2的情况，此时\lambda_1-\lambda_2=k(k为常数)，则限制条件可以表示为下图中的两条直线（对任一样本只对应一条），当出现最小值在矩形区域的左边 \\
（图中情况1）时，L为a点与b点y值的最大值；当出现最小值在矩形区域的右边（图中情况2）时，H为c点与d点y值的最大值，即L = \max\{0, \lambda_2^{old} - \lambda_1^{old}\}, \\
H = \min\{C, C + \lambda_2^{old} - \lambda_1^{old}\}$
