SWA发表于2020年的论文"SWA Object Detection"，文章将Stochastic Weights Averaging（SWA）的思想引入目标检测，将训练好的模型（best or last）额外训练12个epoch，用循环cosine decay，然后把12个模型的参数取平均，得到最终模型，效果是众多检测器在coco上都提升了约1个点

目标检测发展至今，已进入瓶颈期，众多研究者在coco上提升1个点变得很难。

SWA的思想如下图所示，将最后几个checkpoint取平均，能达到更好的最优点。
