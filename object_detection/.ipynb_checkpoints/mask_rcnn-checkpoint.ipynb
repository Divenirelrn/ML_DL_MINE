mask rcnn于2018年的论文"Mask R-CNN"提出，一作是何凯明。mask rcnn主要针对实例分割任务设计，但在目标检测上同样得到了提升。mask rcnn的主要贡献是：在FPN+faster rcnn的基础上，加入一个mask分支，此外提出roi align代替roi pooling，在实例分割任务上得到大幅提升，同时在目标检测上也有提升。作者证明，加上mask分支进行训练，这种多任务训练方式对于目标检测来说会有提升。

mask rcnn整体架构如下，在faster rcnn（+FPN）的基础上，加入mask分支，mask分支采用全卷积网络。


ROI Align<br/>
roi pooling在计算时有两次取整运算，一次是将roi区域映射到特征图上，一次是在映射区域内将映射区域分为n个bins，这两次取整会影响最终的精度。因此作者提出roi align，不用取整，保留小数，利用双线性插值得到poi align的结果。如下图：
