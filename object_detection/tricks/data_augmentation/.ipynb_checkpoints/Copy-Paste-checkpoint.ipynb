copy-paste出自21年论文"Simple Copy-Paste is a Strong Data Augmentation Method
for Instance Segmentation"，提出一种在实例分割中有效的数据集扩增方式，在目标检测中同样有不错的表现。

copy-paste方法如下，从mini-batch中随机选取两张图像，从一张图像中随机选取一个或多个物体，经过随机的flip于scale之后，粘贴到另一种图像的任意位置。


与关注粘贴位置的copy paster方法相比，粘贴到任意位置可以起到同样的效果。

large scale jittering<br/>
标准的random scale是在[0.8, 1.25]的范围内scale，而large scale jittering(LSJ)是在[0.1, 2.0]的范围内变换，这个trick能够很好地提升精度。


self-training copy-paste:self-training指用未标注样本来提升模型的表现，作者通过实验证明，copy-paste方法与self-training联用，能够提升模型效果。

作者通过实验发现，在应用数据集扩增时，延长训练epoch能够很好地提升精度。

长尾数据：copy paste方法在长尾数据集LVIS表现较好。

文中的模型用的是efficientnet+nas-FPN
