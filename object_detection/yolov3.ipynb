yolov3的论文出自2018年，题目为"YOLOv3: An Incremental Improvement"，作者指出，yolov3的论文本质上是引用其他工作来对yolo进行改进，得到更好的性能，主要改进包括：借鉴FPN结构输出3个tensor(每个tensor的每个位置输出3个检测框)，用sigmoid来代替softmax来做类别预测，基础网络采用darknet53(引入shorcut连接)。

yolov3沿用v2的思想，坐标偏移预测相对于网格点的偏移。

class prediction<br/>
class prediction用sigmoid取代softmax，主要是为了应对标签有重叠的问题，如women与person。

prediction across scales<br/>
借鉴FPN的思想，输出3个tensor，输入为256x256的情况下，分别输出32x32、16x16、8x8的三个tensor。不同于FPN的是，上采样的特征图与forward的特征图是cat操作。

k-means聚类<br/>
与yolov2类似，采用聚类算法预设anchors,首先将所有图像resize成输入大小，然后设中心点为9，聚类出9个anchor，分配给3个tensor。
