xception是inception的极限版本，是depth-wise conv在inception中的应用。xception对于输入，先通过一个1x1卷积进行升维，然后用depth-wise conv，即在每一个维度上进行卷积。xception相对于inceptionV3，在imagenet上精度略有提升，在超大数据集上提升明显。它的优势在于对特征的高效利用。

极限版本的inception模块与深度可分离卷积有以下区别：<br/>
1)深度可分离卷积是先有一个depth-wise操作，然后再有一个point-wise操作（1x1卷积）。而极限版本的inception是反过来，先1x1，再depth-wise<br/>
2)深度可分离卷积通常没有非线性变换，而inception包含非线性变换（ReLU）(*)
