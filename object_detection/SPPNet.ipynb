SPPNet于2015年提出，其历史贡献是应用spatial pyramid pooling结构，打破了卷积神经网络对于输入图像尺寸的限制，同时多个尺度的pooling操作，提升了模型对于几何形变的鲁棒性。同时，作者将SPP结构引入RCNN检测网络，只需一次特征提取，然后对任意尺寸的候选区域对应的特征图区域，应用SPP构建一个固定长度的特征向量，极大地减少了训练与测试的时间。

SPPNET结构如下图所示，先经过一次卷积，然后利用SPP将任意尺寸的特征图映射为一个固定长度的特征向量。因此，打破了卷积神经网络对于输入尺寸的限制。


SPP结构如下图所示，SPP的优点如下：<br/>
1)能够产生固定长度的输出，因此对输入尺寸没有限制<br/>
2)多个pooling尺寸，对于物体形变具有更强的鲁棒性，防止过拟合<br/>


SPP可以以一个组件的形式应用于其它网络结构。

多尺度训练：作者采用两种多尺度训练的方式，一是选用两种尺寸,180x180,224x224，其中224x224是先将图像resize为256x256，然后进行裁剪，而180x180是将224x224的图像resize而来，每个epoch变换一次尺寸。另一种是在[180, 224]的范围内随机挑选尺寸，同样是每个epoch变换一次尺寸。

分类trick:作者采用的分类trick包括：由256的图像裁剪为224，10-crop evaluation，multi-scale training,full-image representations(10-crop测试时将整张图像加入进去)

multi-view testing:先将图像resize成256（最小边），然后利用网络提取图像及其horizontal flip的特征，然后在特征图上提取224x224区域对应的特征图区域，作者采用了10 views(4个corner加一个center x2)

SPP应用于检测框架时如下，在实现时将每一个候选区域投影到特征图，然后对于每一个任意尺寸的特征图区域，采用SPP将其处理为一个固定长度的输出。


检测tricks:multi-scale feature extraction（[480,576,688,864,1200]），由于检测区域相对较小，因此采用更大的输入尺寸。细节：微调网络时负样本iou范围为[0.1,0.5],svm训练时同样是0.3，但各个候选框之间的iou不超过0.7.

model combination(ensemble):首先利用网络得到候选区域的socre，然后应用NMS，最后对于每一个proposal，选score值高的作为最终结果。

将proposal投影到feature map:在特征图上获取proposal左上、右下两个点的坐标位置：
$$
left(top): x' = floor(x/S) + 1 \\
right(bottom): x' = ceil(x/S) + 1
$$
$其中，S为下采样倍数的乘积$
