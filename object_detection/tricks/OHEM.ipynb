OHEM( Online Hard Example Mining)出自16年的论文"Training Region-based Object Detectors with Online Hard Example Mining"，是一种region-based目标检测方法中用到的训练trick。在线难例挖掘的思想是：faster rcnn或fast rcnn在训练时，随机采样256个roi区域输入到分类与回归头进行loss计算，256个roi正负样本比例大约是1:3，而OHEM在采样256个样本时，首先将所有的rois输入到分类与回归头进行loss计算，然后挑选loss最大的前256个rois进行loss计算与反向传播。

hard example mining在以前经常用于SVM训练，其方法是：首先维护一个较小的数据集，其中包含较少负样本，固定住数据集进行SVM训练，直至收敛；然后固定住SVM，将所有的负样本输入至SVM进行预测，找出其中分成正类的FP，加入到数据集中，如此反复迭代。一般进行一次就不会再有精度上的提升。

而这种方法对于训练深层网络开销很大，训练周期过长，因此作者提出一种在线难例挖掘。
