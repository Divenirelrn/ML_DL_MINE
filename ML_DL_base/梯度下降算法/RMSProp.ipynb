RMSProp(root mean square prop，均方根传递)，是不同于SGDM的另一种加速收敛过程的优化算法，也就是加速收敛的另一个思路;同时解决了AdaGrad训练后期学习率趋于0的问题。

RMSProp的迭代公式为：<br/>
$$
   (x_k)_i = (x_{k-1})_i + \alpha \frac{(g_k)_i}{\sqrt{s_{(g_k)_i}} + \epsilon}
$$
$其中，s_{(g_k)_i}=\beta s_{(g_{k-1})_i} + (1-\beta)(g_k)_i^2，此处用了指数滑动平均的方式来计算s_{(g_k)_i}的值$

RMSProp的理解：<br/>


假设模型只有两个参数w与b,图中蓝线为标准梯度下降算法的迭代路径，由于每一次前进方向为w与b的方向的组合，因此可以看出，路径在垂直方向震荡较大
$s_{w_k} = \beta s_{w_{k-1}} + (1-\beta)w_k^2$<br/>
$s_{b_k} = \beta s_{b_{k-1}} + (1-\beta)b_k^2$<br/>
由于b的历史值较大，因此$s_{b_k}$较大，则学习率较小，最终计算出的\Delta b较小；<br>
同理，由于w的历史值较小，因此$s_{w_k}$较小，则学习率较大，最终计算出的\Delta w较大；<br>
因此，可以得到图中绿色的迭代路径
