VGG在2014年被提出，其突出特点是网络的统一：全部使用3x3的卷积，并且下采样时filter map size均减少一般，且channel增加一倍。这种统一性使得VGG net作为一种backbone，直至今日都在被广泛使用，尤其是VGG-16与VGG-19。

VGG论文的另一大突出贡献是试验了用小的filter size增加网络深度，在减少参数量的同时，其效果要优于大尺寸的filter（如用3个3x3的卷积代替一个7x7的卷积）。

VGG的显著缺点是参数量太大，难以部署在移动端或FPGA。
