dropblock提出于2018年的论文"DropBlock: A regularization method for convolutional networks"，探究了dropout在卷积层的应用，在voc与coco数据集均取得了1个多点的提升。

dropout一般应用于全连接层而不是卷积层，因为卷积层本来就是稀疏连接，某个区域的激活值是相互关联的，使一个像素点失活，特征同样可以传输到后续网络，因此正则化效果不明显。

dropblock如下图所示，drop掉一个连续的区域，迫使网络去看其他地方的特征。


dropblock受启发于cutout，只是用于特征层。在训练时，初始阶段用一个小的drop ratio，后面慢慢增大（线性减小keep_prob），会取得好的效果，类似于scheduledDropPath（暂时略过）。
