k-means算法是一种经典的聚类算法，其算法流程大致如下：<br/>
1)随机挑选几个聚类中心点；<br/>
2)计算各个点到聚类中心的欧式距离，并把每个点归到离其最近的中心点所对应的簇；<br/>
3)重新计算各个簇的中心点（计算均值即可，中心点不一定是样本点）；<br/>
重复以上2、3步骤，知道到达预设聚类次数或簇不再改变为止。

k-means算法的时间复杂度为o(tkmn)，其中t为预设迭代次数，k为预设的簇数，m为样本点数量，n为每个样本的维数；<br/>
k-means算法的空间复杂度为o((m + k)n)，其中k为预设的簇数，m为样本点数量，n为每个样本的维数

k-means的优点：<br/>
1)思路简单，易于实现，时间复杂度较低<br/>
2)对于近似高斯分布的样本，聚类效果较好；<br/>

k-means的缺点：<br/>
1)对初始值敏感，不同的初始值有不同的聚类效果；<br/>
2)聚类簇数k的选取较麻烦，需要人工设定；<br/>
3)对异常值较敏感，异常值很大程度上影响聚类的效果<br/>

k-means的优化：<br/>
k值的选择：<br/>
手肘法：以SSE作为评价指标，通过选取不同的k值，画出SSE与k之间的关系曲线，则曲线的形状类似手肘，选择转折处的k值即为最佳值（下图中选择k=3）：

Gap statistic方法：<br/>
$$Gap(K) = E(logD_k) - logD_k$$
$其中D_k = \sum_{r=1}^{k}{\frac{1}{2n_r}D_r},D_r = \sum_{i,i'\in C_r}{d_{ii'}}, n_r = |C_r|$<br/>
$E(logD_k)为logD_k的期望，这个数值通过蒙特卡洛方法产生，使得Gap(K)最大的K值即为最佳值：$

通过下游任务确定：通过选择不同的k值，比较下游任务的性能指标，从而确定k值

初始聚类中心的选择：<br/>
k-means++：k-means++的思想就是选择k个距离最远初始化聚类中心，具体步骤如下：<br/>
1)随机选择一个样本点作为第一个聚类中心；<br/>
2)计算每个样本点到最近的聚类中心的距离，选择所有距离中最大的一个作为新的聚类中心；<br/>
重复第2)步，直到找出k个聚类中心

采用核函数：对于非凸的数据分布，可以引入核函数将数据映射到高维空间，在高维空间进行聚类，也被称为核k-means算法。

k-means算法本质是EM算法的应用，同样收敛到局部极小值。
