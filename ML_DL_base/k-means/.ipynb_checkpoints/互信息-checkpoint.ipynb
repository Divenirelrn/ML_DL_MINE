互信息表征了两个概率分布之间的相关性

互信息的定义为：<br/>
$$
I(x,y) = \sum_{x}{\sum_y{p(x,y)ln\frac{p(x,y)}{p(x)p(y)}}}
$$

当两个随机变量相互独立时，联合分布等于边缘分布的乘积，此时互信息为0

如对于下面的概率分布：<br/>

$I(x,y) = 0.2 * ln\frac{0.2}{0.5 * 0.5} + 0.3 * ln\frac{0.3}{0.5 * 0.5} + 0.3 * ln\frac{0.3}{0.5 * 0.5} + 0.2 * ln\frac{0.2}{0.5 * 0.5} = 0.02$

互信息与熵和交叉熵的关系：<br/>
$I(x,y) = \sum_{x}{\sum_{y}{p(x,y)ln(p(x,y))}} - \sum_{x}{\sum_y{p(x,y)lnp(x)}} - \sum_{x}{\sum_{y}{p(x,y)lnp(y)}} = \sum_{x}{\sum_{y}{p(x,y)ln(p(x,y))}} - \sum_{x}{(\sum_y{p(x,y))lnp(x)}} - \sum_{y}{(\sum_{x}{p(x,y))lnp(y)}} = -H(x,y) + H(x) + H(y)$<br/>
