P（Polynominal）问题是指具有多项式时间解($o(n),o(n^2), o(n^3),...$)的问题，P即为多项式英文的首字母（非多项式时间：如$o(2^n)$）<br/>
NP(non-deterministic polynominal)问题是指难以找到多项式时间解，但可以在多项式时间内验证一个可行解（猜出一个可行解）<br/>
NPC（non-deterministic polynominal complete）问题，首先是一个NP问题，其次所有的NP问题都能约化成它，只要解决了NPC问题，那么所有的NP问题也就解决了<br/>
NP难（NP-hard）问题，所有的NP问题都能约化成它，但不一定是NP问题
