spatial transformer network发表于15年，其历史贡献是提出一种空间注意力模块，能够嵌入到卷积网络中，能够对输入图像做类似于仿射变换之类的变换，使得网络具有平移、旋转、缩放等不变性。

STN的作用如下图所示，能够自动对输入或特征图做校正，使得特征更加容易学习。


spatial transformer模块的结构如下，由localisation net、grid generator与sampler组成。其中localisation net通过两个fc学习变换的参数，以仿射变换为例，学习仿射变换的6个参数，grid generator接受产生的参数，产生输入与输出之间的位置对应关系，即grid，sampler通过双线性插值根据grid由输入特征图的值得到输出特征图。
