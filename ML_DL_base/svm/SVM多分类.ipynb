基于binary SVM（二分类SVM），可将SVM用于多分类，一般有两种思路，一种是one-against-one，one-against-all。

one-against-one是一种实际应用效果较好的多分类方法，也是LibSVM库采用的方法。
$给定m个样本的数据集X = \{ (x_1, y_1), ...,(x_m, y_m) \}，x_i \in R^n，类标签y_n={1,2,...,K}$<br/>
$one-against-one的方法是在每两个类之间构造一个binary SVM，以下图(a)为例：$

$一共3类，虚线d_{12}是1类与2类之间的决策边界，虚线d_{13}是1类与3类之间的决策边界，\\虚线d_{23}是2类与3类之间的决策边界$<br/>
$对于新的样本，采用投票法进行分类，例如，以1类与2类决策函数进行分类，若新样本判别为1类，\\则1类加一票，以1类与3类决策函数进行分类......最后得票数最多的类即判别为最终结果$<br/>
$如出现平票的情况，简单地以下标低的类作为判别结果$

one-to-all(也成为one-to-rest)同样也是构造多个决策超平面，每一个决策超平面是划分正类与其余类。对于每一个类，将此类作为+1类，其余类作为-1类。如下图所示：

低i类的决策函数采用：<br/>
$$
d_{new}^i = (w^i)^T\Phi(x_{new}) + b^i
$$
$该决策函数可以看成新数据到各个决策边界的带符号的函数间隔，\\如果新数据属于第i类，那么该值为正$<br/>
对于新样本，K个决策函数给出K个输出，我们选择使得决策函数最大的类作为最终输出：
$$
f(x_{new}) = argmax_{i \in \{1,2,...,K\}}[(w^i)^T\Phi(x_{new}) + b^i]
$$
$即对于一个新样本，每个决策函数给出一个输出，输出最大的即为最终的类$

one-to-rest方法通常面临数据不均衡的问题，此时可修改目标函数的惩罚项来进行调整：<br/>
$$
 \begin{cases}
 \min_{w,b,\xi}{\frac{1}{2}w^Tw + C^+ \sum_{y_i=1}{\xi_i} + C^{-}\sum_{y_i=-1}{\xi_i}}\\
 s.t. y_i[w^T\Phi(x_t) + b] >= 1 - \xi_i, i =1,2,..,m \\
 \xi_i >= 0, i=1,2,...,m
 \end{cases}
$$
$其中C^+和C^-分别为+1类与-1类的惩罚项，可调整惩罚项使得正类的惩罚项更高\\来应对样本不均衡问题$
