resnext的贡献是将group conv引入residual模块，同时借鉴inception结构，构建了一个统一的（各模块结构统一）、多分支的网络。同时，引入一个新的参数维度，叫做"cardinality"（组卷积的group数量），是与depth与width同一级别的参数，通过控制该参数，可以方便地提升网络容量。

VGG与resnet的构建原则是模块堆积，inception的特点是split-transform-merge，但inception不方面调整网络。resnext是基于模块堆积与split-transform-merge，构建的模块如下图所示。resnext的构建原则是简单统一，便于调节网络容量，因此下图右面的各分支结构相同。


resnext模块有三种等价实现，见下图。实现时一般用第三种。


cardinality(基数)，即组卷积的组数，是一个非常重要的参数。调整cardinality来提升精度，比调整width与depth更加高效。

模块模板：基于以下两条原则，构建模块模板：<br/>
1)若特征图尺寸相同，则采用相同超参数<br/>
2)特征图尺寸较少一半，则channel增加为原来的2倍。<br/>
